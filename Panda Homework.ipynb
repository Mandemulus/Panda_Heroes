## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
import pandas as pd
purchase_data_df = pd.read_csv('Resources/purchase_data.csv'
            , header='infer'
            , sep=',')

pd_Age_Purchases=pd.DataFrame()
pd_Age_Purchases = purchase_data_df
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
pd_Age_Purchases["Purchase Summary"] = pd.cut(pd_Age_Purchases["Age"]
                                        , bins, labels=group_names, include_lowest=True)
pd_Age_Purchases.drop('Purchase ID', inplace=True,axis=1)
#pd_Age_Purchases.drop('SN', inplace=True,axis=1)
pd_Age_Purchases.drop('Gender',inplace=True,axis=1)
pd_Age_Purchases.drop('Item ID',inplace=True,axis=1)
pd_Age_Purchases.drop('Item Name',inplace=True,axis=1)
pd_Age_Purchases.drop('Age',inplace=True,axis=1)


pd_Age_Purchases_Summary=pd.DataFrame()
pd_Age_Purchases_Summary = pd_Age_Purchases.groupby(['Purchase Summary']).size().reset_index(name='Purchase Count')
pd_Age_Purchases['Purchase Count'] = pd_Age_Purchases_Summary['Purchase Count']

pd_Age_Purchases_Price=pd.DataFrame()
pd_Age_Purchases_Price = pd_Age_Purchases.groupby(['Purchase Summary'])[['Price']].sum()


#print(pd_Age_Purchases)
#this works
pd_Age_Purchases_Count=pd.DataFrame()
pd_Age_Purchases_Count = pd_Age_Purchases.groupby(['Purchase Summary'])[['SN']].nunique()

#print(pd_Age_Purchases_Count)
pd_Age_Purchases_Summary = pd_Age_Purchases_Summary.merge(pd_Age_Purchases_Count, left_on='Purchase Summary', right_on='Purchase Summary')
#pd_Age_Purchases_Summary = pd_Age_Purchases_Summary.rename(columns={"SN": "Per Person"})

pd_Age_Purchases_Summary = pd_Age_Purchases_Summary.merge(pd_Age_Purchases_Price, left_on='Purchase Summary', right_on='Purchase Summary')
pd_Age_Purchases_Summary = pd_Age_Purchases_Summary.rename(columns={"Purchase Summary": "Age Ranges"})
pd_Age_Purchases_Summary["Average Purchase Price"] = round((pd_Age_Purchases_Summary["Price"] / pd_Age_Purchases_Summary['Purchase Count']),2)

pd_Age_Purchases_Summary["Per Person"] = pd_Age_Purchases_Summary['Age Ranges'].size

pd_Age_Purchases_Summary["Avg Total Purchase per Person"] = pd_Age_Purchases_Summary['Price'] / pd_Age_Purchases_Summary['SN']
pd_Age_Purchases_Summary = pd_Age_Purchases_Summary.rename(columns={"Price": "Total Purchase Value"})

pd_Age_Purchases_Summary.drop('Per Person',inplace=True,axis=1)
pd_Age_Purchases_Summary.drop('SN', inplace=True,axis=1)

#pd_Age_Purchases_Summary
pd_Age_Purchases_Summary.style.format({"Average Purchase Price": "${:,.2f}", "Total Purchase Value": "${:,.2f}",
                                        "Avg Total Purchase per Person": "${:,.2f}"

                        }).hide_index()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
import pandas as pd

purchase_data = pd.read_csv('Resources/purchase_data.csv'
            , header='infer'
            , sep=',')
purchase_data1=pd.DataFrame()
purchase_data1 = pd.read_csv('Resources/purchase_data.csv'
            , header='infer'
            , sep=',')
pd_Top_Spenders_Purchases=pd.DataFrame()

pd_Top_Spenders_Purchases = purchase_data1.groupby(['SN']).size().reset_index(name='Purchase Count')

pd_Age_Purchases_Price=pd.DataFrame()
pd_Age_Purchases_Price = purchase_data1.groupby(['SN'])[['Price']].sum()

pd_Top_Spenders_Purchases = pd_Top_Spenders_Purchases.merge(pd_Age_Purchases_Price, left_on='SN', right_on='SN')
pd_Top_Spenders_Purchases = pd_Top_Spenders_Purchases.rename(columns={"Price": "Total Purchase Value"})
pd_Top_Spenders_Purchases["Average Purchase Price"] = round((pd_Top_Spenders_Purchases["Total Purchase Value"] / pd_Top_Spenders_Purchases['Purchase Count']),2)

# sort by Purchase Count Descending
pd_Top_Spenders_Purchases = pd_Top_Spenders_Purchases.sort_values(by=['Total Purchase Value'], ascending=False)

#pd_Age_Purchases_Price
pd_Top_Spenders_Purchases.head(5).style.hide_index()
pd_Top_Spenders_Purchases.style.format({"Total Purchase Value": "${:,.2f}",
                                        "Average Purchase Price": "${:,.2f}"}).hide_index()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
import pandas as pd
purchase_data1=pd.DataFrame()
purchase_data1 = pd.read_csv('Resources/purchase_data.csv'
            , header='infer'
            , sep=',')

pd_Most_Popular_Purchases=pd.DataFrame()

pd_Most_Popular_Purchases = purchase_data1.groupby(['Item ID', 'Item Name']).size().reset_index(name='Purchase Count')

pd_Most_Popular_Purchases_Price=pd.DataFrame()
pd_Most_Popular_Purchases_Price = purchase_data1.groupby(['Item ID'])[['Price']].sum()

pd_Most_Popular_Purchases = pd_Most_Popular_Purchases.merge(pd_Most_Popular_Purchases_Price, left_on='Item ID', right_on='Item ID')

pd_Most_Popular_Purchases['Purchase Price'] = round((pd_Most_Popular_Purchases['Price'] / pd_Most_Popular_Purchases['Purchase Count']),2)


# sort by Purchase Count Descending
pd_Most_Popular_Purchases = pd_Most_Popular_Purchases.sort_values(by=['Price'], ascending=False)

#pd_Most_Popular_Price

pd_Most_Popular_Purchases.style.format({"Price": "${:,.2f}", "Purchase Price": "${:,.2f}"}).hide_index()
```
